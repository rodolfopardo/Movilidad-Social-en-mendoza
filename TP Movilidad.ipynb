```python
#Filtramos Argentina

argentina = gglobal[gglobal['country_region']== "Argentina"]
```

```python
argentina['sub_region_1'].value_counts()
```

```python
#Filtramos Mendoza

mendoza = argentina[argentina['sub_region_1'] == "Mendoza Province"]
```

```python
#Pasamos los valores de fechas a su correspondiente tipo de dato

mendoza['date'] = pd.to_datetime(mendoza['date'])
```

```python
mendoza.columns
```

```python
#Filtrado de informacion 

mendoza = mendoza[['sub_region_2', 'date', 'retail_and_recreation_percent_change_from_baseline',
                   'grocery_and_pharmacy_percent_change_from_baseline',
                   'parks_percent_change_from_baseline', 
                   'transit_stations_percent_change_from_baseline', 
                   'workplaces_percent_change_from_baseline', 
                   'residential_percent_change_from_baseline']]
```

```python
#DataFrame general de trabajo

mendoza.head()
```

```python
#Cambiamos columnas de nombres

mendoza.columns
```

```python
columnas = ['departamento', 'fecha', 'ocio_y_recreacion', 'supermercados_y_farmcias', 'parques', 'transporte',
           'lugares_de_trabajo', 'zonas_residenciales']
mendoza.columns = columnas
```

```python
#Tiempo y Ocio

sns.lineplot(x = "fecha", y = "ocio_y_recreacion", data = mendoza, estimator = "mean")
plt.title('Ocio y recreación en Mendoza', fontsize = 20)
plt.ylabel('Porcentaje de variación')
plt.xlabel('Fecha')
plt.show()
```

```python
sns.boxplot(mendoza['ocio_y_recreacion'], color = "green")
plt.title("Gráfico de caja para conocer distribución de variable ocio y recreación", fontsize = 20)
plt.xlabel('Porcentaje de variación')
plt.show()
```

```python
mendoza.ocio_y_recreacion.describe()
```

```python
#Filtro Gran Mendoza

gran_mendoza = mendoza[(mendoza.departamento == "Maipú Department") | 
        (mendoza.departamento == 'Guaymallén Department') |
        (mendoza.departamento == "Capital Department") | 
        (mendoza.departamento == "Las Heras Department") | 
        (mendoza.departamento == "Luján de Cuyo Department") | 
        (mendoza.departamento == "Godoy Cruz Department")
       ]
```

```python
#Conociendo variables Gran Mendoza

gran_mendoza.tail()
```

```python
#Grafica Departamentos

sns.lineplot(x = "fecha", y = "ocio_y_recreacion", data = gran_mendoza, hue = "departamento")
plt.title('Ocio y recreación por departamento', fontsize = 20)
plt.ylabel('Porcentaje de variación')
plt.xlabel('Fecha')
plt.show()
```

```python
#Grafica Supermercados

sns.lineplot(x = "fecha", y = "supermercados_y_farmcias", data = gran_mendoza, hue = "departamento")
plt.title('Supermercados y Farmacias', fontsize = 20)
plt.ylabel('Porcentaje de variación')
plt.xlabel('Fecha')
plt.show()
```

```python
#Grafica Supermercados

sns.lineplot(x = "fecha", y = "parques", data = gran_mendoza, hue = "departamento")
plt.title('Parques', fontsize = 20)
plt.ylabel('Porcentaje de variación')
plt.xlabel('Fecha')
plt.show()
```

```python
#Grafica Transporte

sns.lineplot(x = "fecha", y = "transporte", data = gran_mendoza, hue = "departamento")
plt.title('Transporte', fontsize = 20)
plt.ylabel('Porcentaje de variación')
plt.xlabel('Fecha')
plt.show()
```

```python
#Grafica Trabajo

sns.lineplot(x = "fecha", y = "lugares_de_trabajo", data = gran_mendoza, hue = "departamento")
plt.title('Lugares de trabajo', fontsize = 20)
plt.ylabel('Porcentaje de variación')
plt.xlabel('Fecha')
plt.show()
```

```python
#Grafica Residencias

sns.lineplot(x = "fecha", y = "zonas_residenciales", data = gran_mendoza, hue = "departamento")
plt.title('Zonas residenciales', fontsize = 20)
plt.ylabel('Porcentaje de variación')
plt.xlabel('Fecha')
plt.show()
```
